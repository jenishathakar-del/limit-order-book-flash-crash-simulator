```python
import numpy as np

# Baseline market price for a stock is $100.00
fair_value = 100.00

# Generate an order book of buy orders (bids) sitting just below fair value
np.random.seed(42)
bids = fair_value - np.random.exponential(scale=1.5, size=200)
bids = np.sort(bids)[::-1] # Highest buyers at the top

# Generate sell orders (asks) sitting just above fair value
asks = fair_value + np.random.exponential(scale=1.5, size=200)
asks = np.sort(asks) # Lowest sellers at the top

print(f"Order Book Initialized. Tightest Bid: ${bids[0]:.2f} | Tightest Ask: ${asks[0]:.2f}")
print(f"Initial Bid-Ask Spread: ${asks[0] - bids[0]:.2f}")
```

```python
# A sudden macroeconomic scare causes institutional HFT algorithms to cancel their bids
# We wipe out 40% of the highest-priced buy orders sitting in the book
liquidity_drain_index = int(len(bids) * 0.40)
depleted_bids = bids[liquidity_drain_index:]

# A massive market sell order of 100,000 shares hits the depleted book all at once
# The order tears through the remaining shallow bids, driving the price downward
initial_execution_price = bids[0]
crash_execution_price = depleted_bids[40] # The price drops deep into the order book

print("--- FLASH CRASH ALGORITHMIC SIMULATION ---")
print(f"Pre-Crash Top Bid: ${initial_execution_price:.2f}")
print(f"Post-Drain Cleared Bid Price: ${crash_execution_price:.2f}")
print(f"Percentage Flash Devaluation: {((crash_execution_price - initial_execution_price) / initial_execution_price) * 100:.2f}%")
```

```python
price_drop_pct = abs(((crash_execution_price - initial_execution_price) / initial_execution_price) * 100)

print("--- MARKET INTEGRITY REPORT ---")
if price_drop_pct >= 5.0:
    print(f"VOLATILITY HALT TRIGGERED: Price cascaded by {price_drop_pct:.2f}%.")
    print("Action: Order book frozen. Algorithmic circuit breaker deployed successfully.")
else:
    print(f"Market within nominal variance bounds. Price drop: {price_drop_pct:.2f}%.")
```
